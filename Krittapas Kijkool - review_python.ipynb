- Must start with a letter or underscore
- No spaces (use underscores `_` or camelCase)

# 🚦 4. Conditions and Decision Making

Clinical practice is full of decision trees! Python's `if`, `elif`, and `else` let you automate these pathways.

### 🔍 Logical Operators
- `and` → True if BOTH are true
- `or` → True if AT LEAST ONE is true
- `not` → Reverses the boolean

````python
# Example 1: Fever Detection
temp = 38.2

if temp >= 38.0:
    print("⚠️ Fever detected. Recommend further evaluation.")
else:
    print("✅ Temperature is within normal range.")
````

````python
# Example 2: Blood Pressure Classification
sys_bp = 145  # systolic
dia_bp = 92   # diastolic

if sys_bp < 120 and dia_bp < 80:
    bp_status = "Normal"
elif sys_bp < 130 or dia_bp < 80:
    bp_status = "Elevated"
else:
    bp_status = "Hypertension Stage 1/2"

print(f"Blood pressure reading: {sys_bp}/{dia_bp} → {bp_status}")
````

💡 **Expected Output:** `Blood pressure reading: 145/92 → Hypertension Stage 1/2`

````python
# Example 3: Blood Glucose Interpretation
fasting_glucose = 115

if fasting_glucose < 100:
    print("🟢 Normal fasting glucose")
elif 100 <= fasting_glucose < 126:
    print("🟡 Prediabetes range")
else:
    print("🔴 Diabetes diagnostic range (requires confirmation)")
````

````python
# Example 4: Oxygen Saturation Warning System using 'and'/'not'
spo2 = 89
has_lung_disease = True

if not has_lung_disease and spo2 < 95:
    print("🔔 URGENT: Low oxygen saturation. Immediate assessment needed!")
elif has_lung_disease and spo2 >= 88:
    print("📝 Monitoring: COPD patient baseline acceptable.")
else:
    print("📊 Oxygen levels require physician review.")
````

# 🧩 5. Functions

Instead of rewriting code for every patient, we **package it into functions**. This reduces errors, saves time, and standardizes clinical logic.

### 📐 Structure:
```python
def function_name(parameters):
    # do calculations or logic
    return result
```

````python
# Function 1: BMI Calculator
def calculate_bmi(weight, height):
    """Returns BMI given weight in kg and height in meters."""
    return weight / (height ** 2)

# Test it
patient1_bmi = calculate_bmi(68, 1.65)
print(f"Patient 1 BMI: {patient1_bmi:.1f}")
````

````python
# Function 2: Pediatric Medication Dosage (mg)
def calc_child_dose(weight_kg, mg_per_kg):
    """Calculates total dose in milligrams for children."""
    total = weight_kg * mg_per_kg
    # Cap at maximum recommended dose for safety
    max_dose = mg_per_kg * 80  # example cap
    return min(total, max_dose)

dose = calc_child_dose(15, 10) # amoxicillin example
print(f"Child dose calculated: {dose} mg")
````

````python
# Function 3: Temperature Classifier
def classify_temperature(temp):
    if temp < 36.0:
        return "Hypothermic"
    elif 36.0 <= temp <= 37.5:
        return "Normal"
    elif 37.6 <= temp <= 38.3:
        return "Low-grade Fever"
    else:
        return "High Fever"

print(f"38.5°C → {classify_temperature(38.5)}")
print(f"36.4°C → {classify_temperature(36.4)}")
````

````python
# Function 4: Pulse Rate Analyzer
def analyze_pulse(rate, is_resting=True):
    if is_resting:
        if rate < 60:
            return "Bradycardia (Athlete? Check meds)"
        elif rate > 100:
            return "Tachycardia (Pain, fever, anxiety?)"
        else:
            return "Normal resting rate"
    else:
        return "Non-resting rate: contextual evaluation needed"
````

````python
# Calling functions with different patient data
p_rate_1 = 110
p_rate_2 = 74
print(f"Patient A (110 bpm, resting): {analyze_pulse(p_rate_1, True)}")
print(f"Patient B (74 bpm, post-walk): {analyze_pulse(p_rate_2, False)}")
````

💡 **Expected Output:**  
`Patient A (110 bpm, resting): Tachycardia (Pain, fever, anxiety?)`  
`Patient B (74 bpm, post-walk): Non-resting rate: contextual evaluation needed`

# 🔁 6. Iterations and Loops

Clinicians don't manually check one chart at a time when a ward has 20 patients. **Loops** let you automate repetitive tasks across datasets.

- `for` loop: Iterate a known number of times (e.g., through a list)
- `while` loop: Repeat until a condition changes (e.g., monitoring until stable)
- `break`: Exit loop early
- `continue`: Skip current iteration

````python
# Example 1: Loop through a list of patient temperatures
temps = [36.4, 37.1, 38.2, 36.9, 39.0]

for temp in temps:
    status = classify_temperature(temp)
    print(f"Temp: {temp}°C → {status}")
````

````python
# Example 2: Calculate average blood pressure from 5 patients
bp_readings = [118, 125, 132, 110, 140]
total_bp = 0

for reading in bp_readings:
    total_bp = total_bp + reading  # same as total_bp += reading

avg_bp = total_bp / len(bp_readings)
print(f"Clinic average systolic BP: {avg_bp:.1f} mmHg")
````

````python
# Example 3: While loop for simulated patient monitoring
o2_level = 94
minute = 0

print("⏱️ Starting O2 monitoring...")
while o2_level < 96 and minute < 10:
    print(f"Minute {minute}: O2 = {o2_level}% (Below 96% threshold)")
    o2_level += 1  # simulate slow recovery
    minute += 1
print(f"🟢 Stable at minute {minute}. O2 now: {o2_level}%")
````

````python
# Example 4: Using 'break' and 'continue'
glucose_values = [95, 102, 88, 115, 250, 98, 105] # 250 is critical

print("🔍 Scanning glucose levels...")
for val in glucose_values:
    if val > 200:
        print(f"🚨 CRITICAL: {val} mg/dL! Stopping automated scan for clinician review.")
        break  # exit loop immediately
    if val < 70:
        continue  # skip logging normal/hypoglycemia here
    print(f"  -> {val} mg/dL logged.")
````

# 📋 7. Lists and Simple Data Collections

Real clinical data comes in batches: lab panels, patient cohorts, medication lists. Python **lists** (`[]`) are perfect for storing ordered collections.

- Index starts at `0`
- Mutable (can be changed after creation)
- Can hold mixed types (though keeping one type is best practice for data analysis)

````python
# Create a list of patient IDs
patient_ids = ["P001", "P002", "P003", "P004", "P005"]
print(f"All patients: {patient_ids}")
````

````python
# Indexing and Slicing
print(f"First patient: {patient_ids[0]}")      # 0-based index
print(f"Last patient: {patient_ids[-1]}")     # negative index counts from end
print(f"First three: {patient_ids[0:3]}")      # slice (start inclusive, end exclusive)
````

````python
# Modifying lists
patient_ids.append("P006")  # add to end
print(f"After discharge admission: {patient_ids}")

patient_ids.remove("P003")  # remove specific ID
print(f"After discharge: {patient_ids}")

print(f"Total active patients: {len(patient_ids)}") # len() gets length
````

````python
# Parallel lists: names and glucose levels
names = ["Alice", "Ben", "Clara", "David", "Eva"]
glucose = [92, 110, 88, 140, 95]

print("📊 Patient Glucose Report:")
for i in range(len(names)):
    status = "🔴" if glucose[i] >= 126 else "🟢"
    print(f"{status} {names[i]}: {glucose[i]} mg/dL")
````

````python
# Using zip() for cleaner parallel iteration
print("\n✨ Using zip() for paired data:")
for name, glu in zip(names, glucose):
    print(f"{name} -> {glu} mg/dL")
````

# 🏥 9. Simple Challenge Project
## 📊 Patient Vital Sign Monitoring System

You will now combine everything into a mini dashboard simulator. This program:
1. Stores mock patient vital sign data
2. Loops through each patient
3. Evaluates temperature, blood pressure, and heart rate
4. Prints actionable clinical alerts

Run each cell sequentially!

````python
# 📦 Step 1: Set up patient dataset (list of dictionaries)
ward_patients = [
    {"id": "W101", "name": "Maria Gonzalez", "temp": 36.6, "bp": "128/82", "hr": 76},
    {"id": "W102", "name": "Liam Patel", "temp": 38.5, "bp": "150/95", "hr": 112},
    {"id": "W103", "name": "Chen Wei", "temp": 35.8, "bp": "110/70", "hr": 58},
    {"id": "W104", "name": "Sarah Kim", "temp": 37.2, "bp": "135/88", "hr": 84},
    {"id": "W105", "name": "David Smith", "temp": 39.1, "bp": "160/102", "hr": 125}
]
print(f"✅ Loaded data for {len(ward_patients)} patients.")
````

````python
# 🧠 Step 2: Define helper functions for the monitor
def check_temp(t):
    return "Fever" if t >= 38.0 else ("Low" if t < 36.0 else "Normal")

def check_hr(h):
    return "Tachycardia" if h > 100 else ("Bradycardia" if h < 60 else "Normal")

def check_sys_bp(sys_str):
    sys_val = int(sys_str.split("/")[0]) # extract systolic
    if sys_val >= 140:
        return "High"
    return "Normal/Controlled"
````

````python
# 🖨️ Step 3: Run the monitoring loop and print alerts
print("="*45)
print("🏥 WARD VITAL SIGNS MONITORING REPORT")
print("="*45)

for p in ward_patients:
    temp_status = check_temp(p["temp"])
    hr_status = check_hr(p["hr"])
    bp_status = check_sys_bp(p["bp"])
    
    # Determine overall alert level
    alerts = []
    if temp_status == "Fever": alerts.append("🌡️ Fever")
    if temp_status == "Low": alerts.append("❄️ Hypothermia risk")
    if hr_status != "Normal": alerts.append(f"💓 {hr_status}")
    if bp_status == "High": alerts.append("🔺 Elevated BP")
    
    print(f"\n👤 {p['name']} ({p['id']}) | BP: {p['bp']} | HR: {p['hr']}")
    
    if alerts:
        print("⚠️  ALERTS: " + " | ".join(alerts))
        print("📝 Action: Immediate nurse assessment recommended.")
    else:
        print("✅ Status: All vitals within acceptable limits.")
    print("-"*35)
````
